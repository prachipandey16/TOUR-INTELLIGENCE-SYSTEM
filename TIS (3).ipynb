```python
# Stores the data 
df=pd.read_csv('locations.csv',sep=',')
# Shows the data 
df.head()
```

```python
df
```

```python
country= 'India'
city_names = df['location']

longitude = []
latitude = []
geolocator = Nominatim()

for c in city_names.values:
    location = geolocator.geocode(c+ ',' + country)
    latitude.append(location.latitude)
    longitude.append(location.longitude)
```

```python
df['Latitude'] = latitude
df['Longitude'] = longitude
```

```python
l2 = df.iloc[:,-1:-3:-1]
l2
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit(l2)

identified_clusters = kmeans.fit_predict(l2)
identified_clusters = list(identified_clusters)
```

```python
identified_clusters
```

```python
df['Alloted Cluster'] = identified_clusters
```

```python
df
```

```python
input_city = input("Enter a city name: ")
cluster = df.loc[df['location'] == input_city,'Alloted Cluster']
cluster = cluster.iloc[0]
cities = df.loc[df['Alloted Cluster'] == cluster, 'location']
for c in range(len(cities)):
    if cities.iloc[c] == input_city:
        continue
    else:
        print(cities.iloc[c])
```
